```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
from datetime import datetime
```

```python
covid_df = pd.read_csv("covid_19_india.csv")
covid_df.head(10)
```

```python
covid_df.info()
```

```python
covid_df.describe()
```

```python
vaccine_df = pd.read_csv("covid_vaccine_statewise.csv")
vaccine_df.head()
```

```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis=1)
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
covid_df.head()
```

```python
#Active cases

covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values= ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc=max)
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
statewise = statewise.sort_values(by="Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap = "terrain")
```

```python
#Most active States (Top 10)

most_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by=['Active_cases'], ascending=False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with Most Active cases in India", size = 25)
ax = sns.barplot(data = most_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'black')
plt.xlabel("States / UnionTerritory")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#States with Most Deaths (Top 10)

most_death_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by=['Deaths'], ascending=False).reset_index()
fig = plt.figure(figsize=(18,7))
plt.title("Top 10 States with Most Death cases in India", size = 25)
ax = sns.barplot(data = most_death_cases.iloc[:12], y = "Deaths", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'black')
plt.xlabel("States / UnionTerritory")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
#Growth Trend

fig = plt.figure(figsize = (12,6))

ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x = 'Date', y = 'Active_cases', hue='State/UnionTerritory')

ax.set_title("Top 10 Most affected States in India", size = 16)
```

```python
vaccine = vaccine_df.drop(columns=['Sputnik V (Doses Administered)', 'AEFI', '18-44 Years (Doses Administered)', '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'], axis=1)
vaccine.head()
```

```python
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns = {'Total Individuals Vaccinated' : 'Total'}, inplace = True)
vaccine.head()
```

```python
#Most vaccinated State

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated States in India", size = 20)
ax = sns.barplot(data = most_active_cases.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = 'black')
plt.xlabel("States / UnionTerritory")
plt.ylabel("Vaccination")
plt.show()
```
